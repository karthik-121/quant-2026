```python
import numpy as np
from matplotlib import pyplot as plt

x_vals = np.linspace(-10,10,2100)
func = f(x_vals)
first_der = derivatives.numerical_derivative(f,x_vals)
second_der = derivatives.second_derivative(f,x_vals)
third_der = derivatives.third_derivative(f, x_vals)

plt.plot(x_vals,func)
plt.plot(x_vals, first_der, label="first")
plt.plot(x_vals, second_der, label = "second")
plt.plot(x_vals, third_der, label = "third")
plt.legend()
plt.show()

```
