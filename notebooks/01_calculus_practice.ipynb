```python
import sys
from pprint import pprint
from pathlib import Path
import numpy as np


```

```python
# Get the path of the current notebook directory (notebooks)
notebook_dir = Path.cwd()

# Define the path to the 'src' directory (sibling to 'notebooks')
src_dir = notebook_dir.parent / 'src'

# Add the source directory to sys.path if it's not already present
if str(src_dir) not in sys.path:
    sys.path.append(str(src_dir))
    print(f"Added {src_dir} to sys.path")

from quant_math.derivatives import numerical_derivative
import math

def f(x):
    return x**2

def g(x):
    return np.sin(x)

def h(x):
    return np.e**x

def k(x):
    return np.log(x)

print(numerical_derivative(f,3))
print(numerical_derivative(g,0))
print(numerical_derivative(h,1))
print(numerical_derivative(k,2))
```

```python
from quant_math.derivatives import second_derivative

print(second_derivative(f,7))
print(second_derivative(g,0))
print(second_derivative(h,0))
```

```python
from matplotlib import pyplot as plt
import numpy as np

input = np.linspace(-10,10,210)
output = f(input)
output2 = numerical_derivative(f,input)
plt.figure(1)
plt.plot(input,output)
plt.plot(input,output2)

plt.ylim(-100,100)

print(plt.show)




```

```python
sininput = np.linspace(-4*(math.pi), 4*(math.pi), 1000)
sinoutput = g(sininput)
sinoutput2 = numerical_derivative(g,sininput)
plt.figure(2)
plt.plot(sininput,sinoutput, label='function')
plt.plot(sininput,sinoutput2, label = 'derivative')
plt.legend()
plt.show()
```

```python
lninput = np.linspace(-10,10,500)
lnoutput = k(lninput)
lnoutput2 = numerical_derivative(k,lninput)
plt.figure(3)
plt.plot(lninput,lnoutput, label='function')
plt.plot(lninput, lnoutput2, label = 'derivative')
plt.legend()
plt.ylim(-10,10)
plt.show()
```
